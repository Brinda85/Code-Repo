```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Univariate Analysis

```python
df['STATE'].value_counts()
```

```python
sns.countplot(x=df['STATE'])
```

```python
df['STATE'].value_counts().plot(kind='bar',
                                xlabel='STATE',
                                ylabel='Counts',
                                title='Countplot for STATE')
```

```python
for i in con:
    sns.histplot(data=df,x=i,hue='STATE',kde=True)
    plt.title(f'Histoplot for {i}')
    plt.show()
```
